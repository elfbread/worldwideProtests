```python
import pandas as pd
import numpy as np
```

```python
dfprot = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQfb9_QNrIe9LQmQgnGbDJhrVsOPHGDgtYIv0KOhoZYY0nEgIGZQBxbdXYCyNlf96m9kRm46d0j3x5b/pub?output=csv')
dfhappy = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQ6hqAD77S_LJoC6Dud7poNg9-xG_feF5QMYWb1zlwE8Wlt5iTvkB2AU7DX0u6VmXwV6iaICbylhqeq/pub?output=csv')
```

```python
df = dfprot.merge(dfhappy,
                            how='left',
                            left_on='Country',
                            right_on='Country name')
df.head()
```

I calculated the range in the ladder values (upper whisker - lower whisker) to find the full range of values for each country. Then the results of this calculation were appended to the dataframe. df.head() was used to confirm an additional column was added.

```python
# range is difference between upper whisker and lower whisker
Ladder_Range = df.upperwhisker - df.lowerwhisker
df['Ladder Range'] = Ladder_Range
df.head()
```

I want to rename some of the columns in the data set. Some of them include '?' which I do not like.

```python
df.rename(columns = {'Economic motivation?':'Economic_motivation', 'Political motivation?':'Political_motivation', 'Corruption motivation?':'Corruption_motivation'}, inplace=True)
```

There is an issue where instead of a boolean value, some of the columns just have an "X" for True. 

```python
# print column names to replace 'X' with boolean
print(df.columns)
```

```python
#  recode 'X' to boolean where 'X' = 'True' and fillna values with 'False'
df_copy = df.copy()
df_copy['Active protests'] = df_copy['Active protests'].map({'X': True})
df_copy['Economic_motivation'] = df_copy['Economic_motivation'].map({'X': True})
df_copy['Political_motivation'] = df_copy['Political_motivation'].map({'X': True})
df_copy['Corruption_motivation'] = df_copy['Corruption_motivation'].map({'X': True})
df_copy['Protests with a significant outcome'] = df_copy['Protests with a significant outcome'].map({'X': True})
df_copy['Large protests (Over 100,000 protesting)'] = df_copy['Large protests (Over 100,000 protesting)'].map({'X': True})
df_copy['Violent government response'] = df_copy['Violent government response'].map({'X': True})
df_copy['Long protests (more than 3 months)'] = df_copy['Long protests (more than 3 months)'].map({'X': True})
df_copy['Coronavirus-related protests'] = df_copy['Coronavirus-related protests'].map({'X': True})
df_copy.fillna('False', inplace=True)
```
